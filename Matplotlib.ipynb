```python
import matplotlib.pyplot as plt
```

```python
y=2*x
```

```python
a=np.linspace(0,10,10)
```

```python
b=a**3
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.set_title('Big Plot')
axes1.set_xlim(0,10)
axes1.set_ylim(0,800)
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.plot(a,b)
axes2=fig.add_axes([0.2,0.2,.3,.3])
axes1.plot(a,b)
axes2.set_title('Small Plot')
axes2.plot(x,y)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,x**4)
axes[1][1].plot(x,x**5)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,x**4)
axes[1][1].plot(x,x**5)
fig.subplots_adjust(wspace=2)
fig.suptitle('my sub plot')
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y')
mon.plot(a,b,label='a and b')
mon.legend()
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y')
mon.plot(a,b,label='a and b')
mon.legend(loc=4)#location code
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y')
mon.plot(a,b,label='a and b')
mon.legend(loc=(.5,.5))#manual location
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y',color='purple')
mon.plot(a,b,label='a and b',color='green')
mon.legend()
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y',color='#678008')
mon.plot(a,b,label='a and b',color='#C2E924')#RGB Hexcode
mon.legend()
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y',color='red',lw=10)#lw for line width 10 means 10 times default value (linewidth also used for this)
mon.plot(a,b,label='a and b',color='#C2E924')#RGB Hexcode
mon.legend()
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y',color='red',lw=2,ls='--')#ls for line style '--','-',':','-.' (linestyle also used inplace of ls)
mon.plot(a,b,label='a and b',color='#C2E924')#RGB Hexcode
mon.legend()
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y',color='red')
lines=mon.plot(a,b,label='a and b',color='#C2E924')#RGB Hexcode
lines[0].set_dashes([5,3,10,2])#cutomize dashes in line
mon.legend()
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y',color='red')
mon.plot(a,b,label='a and b',color='#C2E924',marker='o')
mon.legend()
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y',color='red')
mon.plot(a,b,label='a and b',color='#C2E924',marker='*')
mon.legend()
```

```python
fig=plt.figure()
mon=fig.add_axes([0,0,1,1])
mon.plot(x,y,label='x and y',color='red')
mon.plot(a,b,label='a and b',color='#C2E924',marker='*',markersize=10,markerfacecolor='red')
mon.legend()
```
